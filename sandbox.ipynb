```python
from utils import show_images
import numpy as np, matplotlib.pyplot as plt
import cv2, csv
```

```python

labels = np.genfromtxt('data/labels.csv',delimiter=',')
print(labels)
```

```python
def read_csv(file):
    with open(file,'r') as f:
        reader = csv.reader(f,delimiter=',')
        for row in reader:
            yield row
```

```python

labels=(read_csv('data/labels.csv'))
images_top = []
images_bottom = []
for time,mon in labels:
    ts = str(time)
    file='data/img/%s_eye_0.png'%ts
    #print("reading image ",file)
    img = cv2.imread(file)
    if mon=='HDMI-1-1':
        images_top.append(img)
    else:
        images_bottom.append(img)
    
print("top img len",len(images_top),'bottom img len',len(images_bottom))
```

```python
show_images(images_top[70:100])
show_images(images_bottom[70:100])
```
